```python
#### building logic for one word in description
def filter_desc_by_one_keyword(df_3,it_tech_pattern,case):
    df_4=df_3.dropna(subset=['Job Desception'])
    filtered_df= df_4[df_4['Job Desception'].str.contains(it_tech_pattern, case)]
    it_keywords=['computer','Computer','information technology','Information Technology','IT','software',
             'Software','network','Network','cybersecurity','Cybersecurity','developer',
             'Developer','programming','Programming','virtual','Virtual','Data','data','Digital','digital','AI','web']
    IT_jobs_IDs=[]
    n_IT_jobs=0
    loc=0
    for job_desc in filtered_df['Job Desception'].iloc[:]:
        flag=0
        for keyword in it_keywords:
            try:
                if keyword in job_desc:
                    flag=1
            except TypeError:
            #print('null value')
                pass
        if flag==1:
            n_IT_jobs=n_IT_jobs+1
            IT_jobs_IDs.append(filtered_df['Job ID'].iloc[loc])
            # filtered_df.iloc[5]['Job ID'] and filtered_df['Job ID'].iloc[5] are similar
        
        loc=loc+1
    
    n_IT_jobs
    len(IT_jobs_IDs)
    filtered_df_desc=filtered_df[filtered_df['Job ID'].isin(IT_jobs_IDs)]
    
    print(f'length of filtered_df_desc:{len(filtered_df_desc)}\n\n')
    print(f'Unique values in filtered_df_desc:\n{filtered_df_desc["Job Title"].unique()}\n\n')
    print(f"Number of Unique values in filtered_df_desc:{len(filtered_df_desc['Job Title'].unique())}\n\n")
    
    if len(filtered_df_desc)==len(filtered_df):
        print('It is nearly 100 % varified by Jobs desc that It belongs to IT and Tech Jobs\n\n')
    else:
        percent=len(filtered_df_desc)/len(filtered_df)*100
        print(f'It is nearly {percent} % varified by Jobs desc that It belongs to IT and Tech Jobs\n\n')
        
    defferences=[] # this is list of job titles present in filtered_df_title but not in filtered_df_desc
    defferences1=[] # List of titles that are not cleared as IT and Tech Job due to abnoramal desc
    defferences2=[] # List of titles that are not cleared as IT and Tech Job due to not matching of II_keywords in desc
    for value in filtered_df['Job Title'].unique():
        if value not in filtered_df_desc["Job Title"].unique():
            defferences.append(value)
            if filtered_df[filtered_df['Job Title']==value]['Job Desception'].isnull().sum()==1:
               defferences1.append(value)
            elif len(str(filtered_df[filtered_df['Job Title']==value]['Job Desception']))<64:
               defferences1.append(value)
            else:
                defferences2.append(value)
                
            
   
    print(f'list of {len(defferences)} job titles which present in filtered_df_title but not present in filtered_df_desc:\n{defferences}\n\n')
    print(f'List of {len(defferences1)} titles that are not cleared as IT and Tech Job due to abnoramal desc:\n{defferences1}\n\n')
    print(f'List of  {len(defferences2)} titles that are not cleared as IT and Tech Job due to not matching of II_keywords in desc:\n{defferences2}\n\n')
```

```python
def filter_desc(filtered_df_title,n_abnormal_desc):
    it_keywords=['computer','Computer','information technology','Information Technology','IT','software',
             'Software','network','Network','cybersecurity','Cybersecurity','developer',
             'Developer','programming','Programming','virtual','Virtual','Data','data','Digital','digital','AI','web']
    IT_jobs_IDs=[]
    n_IT_jobs=0
    loc=0
    for job_desc in filtered_df_title['Job Desception'].iloc[:]:
        flag=0
        for keyword in it_keywords:
            try:
                if keyword in job_desc:
                    flag=1
            except TypeError:
            #print('null value')
                pass
        if flag==1:
            n_IT_jobs=n_IT_jobs+1
            IT_jobs_IDs.append(filtered_df_title['Job ID'].iloc[loc])
            # filtered_df.iloc[5]['Job ID'] and filtered_df['Job ID'].iloc[5] are similar
        
        loc=loc+1
    
    n_IT_jobs
    len(IT_jobs_IDs)
    filtered_df_desc=filtered_df_title[filtered_df_title['Job ID'].isin(IT_jobs_IDs)]
    
    print(f'length of filtered_df_desc:{len(filtered_df_desc)}\n\n')
    print(f'Unique values in filtered_df_desc:\n{filtered_df_desc["Job Title"].unique()}\n\n')
    print(f"Number of Unique values in filtered_df_desc:{len(filtered_df_desc['Job Title'].unique())}\n\n")
    
    if len(filtered_df_desc)+n_abnormal_desc==len(filtered_df_title):
        print('It is nearly 100 % varified by Jobs desc that It belongs to IT and Tech Jobs\n\n')
    else:
        percent=(len(filtered_df_desc)+n_abnormal_desc)/len(filtered_df_title)*100
        print(f'It is nearly {percent} % varified by Jobs desc that It belongs to IT and Tech Jobs\n\n')
        
    defferences=[] # this is list of job titles present in filtered_df_title but not in filtered_df_desc
    defferences1=[] # List of titles that are not cleared as IT and Tech Job due to abnoramal desc
    defferences2=[] # List of titles that are not cleared as IT and Tech Job due to not matching of II_keywords in desc
    for value in filtered_df_title['Job Title'].unique():
        if value not in filtered_df_desc["Job Title"].unique():
            defferences.append(value)
            if filtered_df_title[filtered_df_title['Job Title']==value]['Job Desception'].isnull().sum()==1:
               defferences1.append(value)
            elif len(str(filtered_df_title[filtered_df_title['Job Title']==value]['Job Desception']))<64:
               defferences1.append(value)
            else:
                defferences2.append(value)
                
            
   
    print(f'list of {len(defferences)} job titles which present in filtered_df_title but not present in filtered_df_desc:\n{defferences}\n\n')
    print(f'List of {len(defferences1)} titles that are not cleared as IT and Tech Job due to abnoramal desc:\n{defferences1}\n\n')
    print(f'List of  {len(defferences2)} titles that are not cleared as IT and Tech Job due to not matching of II_keywords in desc:\n{defferences2}\n\n')
```
